```python
# Filter the dataframe by genre
movies1 = movies[(movies.Genre == 'action') | (movies.Genre == 'adventure') | (movies.Genre == 'animation')
                 | (movies.Genre == 'comedy') | (movies.Genre == 'drama')]
```

```python
movies1.head(10)
```

```python
# Filter the dataframe by studio
movies2 = movies1[(movies1.Studio == 'Buena Vista Studios') | (movies1.Studio == 'Fox') | (movies1.Studio == 'Paramount Pictures')
                  | (movies1.Studio == 'Sony') | (movies1.Studio == 'Universal') | (movies1.Studio == 'WB')]
```

```python
movies2.head()
```

```python
len(movies2)
```

```python
print(movies2.Genre.unique())
print(movies2.Studio.unique())
print(len(movies2))
movies2.columns
```

```python
# Define the style
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Plot the boxsplots
ax = sns.boxplot(data=movies2, x='Genre', y='Gross % US', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Add in points to show each observation
sns.stripplot(x='Genre', y='Gross % US', data=movies2, jitter=True, size=6, linewidth=0, hue = 'Studio', alpha=0.7)

ax.axes.set_title('Domestic Gross % by Genre',fontsize=25)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross % US',fontsize=20)

# Define where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
