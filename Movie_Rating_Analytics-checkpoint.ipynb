```python
g = sns.FacetGrid(movies, row = 'Genre', col = 'Year', hue = 'Genre')
keys = dict(s =50, linewidth = 0.5, edgecolor = 'black')
g = g.map(plt.scatter, 'CriticsRating', 'AudienceRating', **keys)
```

```python
g = sns.FacetGrid(movies, row = 'Genre', col = 'Year', hue = 'Genre')
keys = dict(s =50, linewidth = 0.5, edgecolor = 'black')
g = g.map(plt.scatter, 'CriticsRating', 'AudienceRating', **keys)
g.set(xlim = (0, 100), ylim = (0, 100))
for ax in g.axes.flat: # creating diagonals
    ax.plot((0,100), (0, 100), c = 'gray', ls = "--")
g.add_legend()
```

Dashboard

```python
f, axes = plt.subplots(3, 2, figsize = (15,12))
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, ax = axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticsRating, ax = axes[0,1])
k3 = sns.violinplot(data = movies, x = 'Year', y = 'BudgetMillions', ax =axes[1,0])
k4 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, shades = True, shade_lowest = False, cmap = "Reds", ax =axes[1,1])
axes[2,0].hist(movies.AudienceRating, bins = 10)
axes[2,1].hist(movies.CriticsRating, bins = 10)
plt.show()
```

```python
sns.set_style('dark', {'axes.facecolor':'black'})
f, axes = plt.subplots(3, 2, figsize = (15,12))
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, ax = axes[0,0], shade = True, 
                 shade_lowest = True, cmap = 'inferno')
k1b = sns.kdeplot(movies.BudgetMillions, movies.AudienceRating, ax = axes[0,0], cmap = 'cool', )
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticsRating, ax = axes[0,1], shade = True, 
                 shade_lowest = True, cmap = 'inferno')
k2b = sns.kdeplot(movies.BudgetMillions, movies.CriticsRating, ax = axes[0,1], cmap = 'cool')
k3 = sns.violinplot(data = movies, x = 'Year', y = 'BudgetMillions', ax =axes[1,0], palette = 'YlOrRd')
k4 = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, ax =axes[1,1], shade = True, 
                 shade_lowest = False, cmap = "Blues_r")
k4b = sns.kdeplot(movies.CriticsRating, movies.AudienceRating, ax =axes[1,1], cmap = "gist_gray_r")
axes[2,0].hist(movies.AudienceRating, bins = 10, color = 'Brown')
axes[2,1].hist(movies.CriticsRating, bins = 10, color = 'Brown')
plt.show()
```

```python
list1 = []
labels1 = []
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    labels1.append(gen)

sns.set_style("whitegrid")
fig, ax = plt.subplots()
vis4 = plt.hist(list1, bins = 15, stacked = True, rwidth = 1, label =labels1)
fig.set_size_inches(10,8)
plt.title("Movies Budget Distribution", fontsize = 25, color ="Green")
plt.xlabel("BudgetMillions", color = "Blue", fontsize = 18)
plt.ylabel("No. of Movies", color = "Red", fontsize = 18)
plt.legend(frameon = True,fancybox = True, shadow = True, framealpha = 1, prop = {"size":15})
plt.show()
```
